```python
import ROOT
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import pandas as pd

aero_thetac_results ={}
aero_thetac_res_results ={}
gas_thetac_results ={}
gas_thetac_res_results ={}


file_count =0
folder_path = "/drive-0/eic-old/ana"
aero_gas_keys ={"theta_dist_Aerogel":0,"thetaResid_dist_Aerogel":1,"theta_dist_Gas":2,"thetaResid_dist_Gas":3}

columns = [
    "nr", "analysis_date", "doneby", "file", "pid", "mom", "eta_bin", "eta",
    "aero_thetac_hist_entries", "aero_thetac_hist_mean", "aero_thetac_hist_rms",
    "aero_thetac_fit_mean", "aero_thetac_fit_mean_err", "aero_thetac_fit_sigma", "aero_thetac_fit_sigma_err", "aero_thetac_fit_const", "aero_thetac_fit_const_err",
    "aero_thetac_fit_chisq", "aero_thetac_fit_ndf",
    "aero_thetac_res_hist_entries", "aero_thetac_res_hist_mean","aero_thetac_res_hist_rms", 
    "aero_thetac_res_fit_mean", "aero_thetac_res_fit_mean_err", "aero_thetac_res_fit_sigma", "aero_thetac_res_fit_sigma_err", "aero_thetac_res_fit_const", "aero_thetac_res_fit_const_err",
    "aero_thetac_res_fit_chisq", "aero_thetac_res_fit_ndf",
    "gas_thetac_hist_entries", "gas_thetac_hist_mean", "gas_thetac_hist_rms",
    "gas_thetac_fit_mean", "gas_thetac_fit_mean_err", "gas_thetac_fit_sigma", "gas_thetac_fit_sigma_err", "gas_thetac_fit_const", "gas_thetac_fit_const_err",
    "gas_thetac_fit_chisq", "gas_thetac_fit_ndf",
    "gas_thetac_res_hist_entries", "gas_thetac_res_hist_mean","gas_thetac_res_hist_rms",
    "gas_thetac_res_fit_mean", "gas_thetac_res_fit_mean_err", "gas_thetac_res_fit_sigma", "gas_thetac_res_fit_sigma_err", "gas_thetac_res_fit_const", "gas_thetac_res_fit_const_err",
    "gas_thetac_res_fit_chisq", "gas_thetac_res_fit_ndf"
]
df = pd.DataFrame(columns=columns)

def process_histogram(o, prefix):
    if o.GetEntries()==0 or (o.GetMean()==0 and o.GetRMS()==0):
        return {
            f"{prefix}_hist_entries": o.GetEntries(),
            f"{prefix}_hist_mean": o.GetMean(),
            f"{prefix}_hist_rms": o.GetRMS(),
            f"{prefix}_fit_mean_err": -999,
            f"{prefix}_fit_sigma": -999,
            f"{prefix}_fit_sigma_err": -999,
            f"{prefix}_fit_const": -999,
            f"{prefix}_fit_const_err": -999,
            f"{prefix}_fit_chisq": -999,
            f"{prefix}_fit_ndf": -999
        }
    else:
      xxmin = o.GetBinCenter(o.GetMaximumBin ())-2*o.GetRMS()
      xxmax = o.GetBinCenter(o.GetMaximumBin ())+2*o.GetRMS()
      r=o.Fit("gaus","WSQ","",xxmin,xxmax)
      
      return {
          f"{prefix}_hist_entries": o.GetEntries(),
          f"{prefix}_hist_mean": o.GetMean(),
          f"{prefix}_hist_rms": o.GetRMS(),
          f"{prefix}_fit_mean": r.Parameter(1),
          f"{prefix}_fit_mean_err": r.ParError(1),
          f"{prefix}_fit_sigma": r.Parameter(2),
          f"{prefix}_fit_sigma_err": r.ParError(2),
          f"{prefix}_fit_const": r.Parameter(0),
          f"{prefix}_fit_const_err": r.ParError(0),
          f"{prefix}_fit_chisq": r.Chi2(),
          f"{prefix}_fit_ndf": r.Ndf()
      }
    



for root, dirs, files in os.walk(folder_path):
    for file in files:
      if file.endswith(".root"):
        file_count += 1
print(f"Total files found in {folder_path}: {file_count}")


fileN=0
for root, dirs, files in os.walk(folder_path):
  for file in files:
    if file.endswith(".root"):
    #   file = "ana_211_5_18.root"
      print(f"Analysing file {fileN+1} of {file_count}: {file}")
      c1 = ROOT.TCanvas(f"{file}")
      c1.Draw()
      c1.Divide(2,2)
      parts = file.split("_")
      pid = int(parts[1])
      mom = float(parts[2])
      eta_bin = int(parts[3].split(".")[0])
      etas=[1.5,1.6,1.7,1.8,1.9,2,2.1,2.2,2.3,2.4,2.5,2.6,2.7,2.8,2.9,3,3.1,3.2,3.3,3.4,3.5]
      eta = etas[eta_bin]


      f = ROOT.TFile(f"/drive-0/eic-old/ana/{file}")
      # print("Objects in the file:")
      for key in f.GetListOfKeys():
        obj = key.ReadObj()
        keys = obj.GetListOfKeys()
        # print(obj.GetName(), obj.ClassName())
        for key in keys:
          obj = key.ReadObj()
          # print(key.GetName(), key.ClassName())
          objs =[]
          if obj.GetName() in aero_gas_keys:
            objs.append(obj)
            for o in objs:
            #   print(f"Object details: {o.GetName()} ({o.ClassName()})")
              if o.GetName() == "theta_dist_Aerogel":
                c1.cd(1)
                o.Draw()
                aero_thetac_results = process_histogram(o, "aero_thetac")
              if o.GetName() == "thetaResid_dist_Aerogel":
                c1.cd(2)
                o.Draw()
                aero_thetac_res_results = process_histogram(o, "aero_thetac_res")
              if o.GetName() == "theta_dist_Gas":
                c1.cd(3)
                o.Draw()
                gas_thetac_results = process_histogram(o, "gas_thetac")
              if o.GetName() == "thetaResid_dist_Gas":
                c1.cd(4)
                o.Draw()
                gas_thetac_res_results = process_histogram(o, "gas_thetac_res")
            
      # print("appending")
      # print(**aero_thetac_results)
      new_row = {
      'nr': fileN,
      'analysis_date': '2023-01-02',
      'doneby': 'Deepak Samuel',
      'file': file,
      'pid': pid,
      'mom': mom,
      'eta_bin': eta_bin,
      'eta': eta,
      **aero_thetac_results,
      **aero_thetac_res_results,
      **gas_thetac_results,
      **gas_thetac_res_results
      }
      c1.Draw()
      new_df = pd.DataFrame([new_row])
      df = pd.concat([df, new_df], ignore_index=True)
      fileN= fileN+1
    #   break
  print("Done...")
```
